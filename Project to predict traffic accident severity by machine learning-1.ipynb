```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('C:\Books and Infor on Materials\Data Science\Applied data science Capstone\Data-Collisions(2).csv')
df.describe()
```

```python
sns.countplot(df['SEVERITYCODE'],data=df)
```

```python
numerical=['PERSONCOUNT','PEDCOUNT','PEDCYLCOUNT','VEHCOUNT']
categorical=['SEVERITYDESC','COLLISIONTYPE','JUNCTIONTYPE','WEATHER','ROADCOND','LIGHTCOND']
df=df[numerical+categorical]
df.shape
```

```python
fig,ax=plt.subplots(1,7,figsize=(30,5))
for variable,subplot in zip(categorical,ax.flatten()):
   sns.countplot(df[variable],ax=subplot)
   plt.subplots_adjust(bottom=0.3,top=0.8)
   for label in subplot.get_xticklabels():
      label.set_rotation(90)
```

```python
corelation=df.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```
